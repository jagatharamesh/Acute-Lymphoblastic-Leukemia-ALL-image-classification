```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import zipfile
import os
```

```python

#read image
src = cv2.imread('WBC-Malignant-Early-001.jpg', cv2.IMREAD_UNCHANGED)
print(src.shape)
#extract green channel
green_channel = src[:,:,1]
#creating threshhold
ret, thresh = cv2.threshold(green_channel,150, 255, cv2.THRESH_BINARY)
backtorgb = cv2.cvtColor(thresh,cv2.COLOR_GRAY2RGB)
print("backtorgb",backtorgb.shape)
# preforming bitwise AND
bitwise_AND = cv2.bitwise_and(src, backtorgb)
print(bitwise_AND.shape)
```

```python
figure_size = 20
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,5,1),plt.imshow(src)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(1,5,2),plt.imshow(green_channel)
plt.title('green_channel'), plt.xticks([]), plt.yticks([])
plt.subplot(1,5,3),plt.imshow(thresh)
plt.title('thresh'), plt.xticks([]), plt.yticks([])
plt.subplot(1,5,4),plt.imshow(backtorgb)
plt.title('backtorgb'), plt.xticks([]), plt.yticks([])
plt.subplot(1,5,5),plt.imshow(bitwise_AND)
plt.title('bitwise_AND'), plt.xticks([]), plt.yticks([])
plt.show()
```
